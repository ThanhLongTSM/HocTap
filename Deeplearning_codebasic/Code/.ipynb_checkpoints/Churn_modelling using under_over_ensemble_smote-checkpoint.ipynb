```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,classification_report
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('E:/AI/Deeplearning_codebasic/Data/Churn_Modelling.csv')
df.sample(5)
```

```python
df = df.drop(['RowNumber','CustomerId','Surname'],axis = 'columns')
```

```python
df['Geography'].replace({'France':0,'Spain':1,'Germany':2},inplace = True)
df['Gender'].replace({'Female':0,'Male':1},inplace = True)
```

```python
col_to_scaler = ['CreditScore','Age','Tenure','Balance','EstimatedSalary','NumOfProducts']
scaler = MinMaxScaler()
df[col_to_scaler] = scaler.fit_transform(df[col_to_scaler])
```

```python
df.Exited.value_counts()
```

### ANN function

```python
def ANN(X_train,y_train,X_test,y_test,loss):
    model = keras.Sequential([
        keras.layers.Dense(10,input_dim=10,activation = 'relu'),
        keras.layers.Dense(5,activation = 'relu'),
        keras.layers.Dense(1,activation = 'sigmoid')
    ])
    model.compile(loss = loss, metrics=['accuracy'],optimizer = 'adam')
    model.fit(X_train,y_train,epochs=100)
    y_preds = model.predict(X_test)
    y_preds = np.round(y_preds)
    print(classification_report(y_test,y_preds))
    return y_preds
```

### Undersampling

```python
count_0,count_1 = df.Exited.value_counts()
df_values_0 = df[df.Exited==0]
df_values_1 = df[df.Exited==1]
```

```python
df_values_0_under = df[df.Exited==0].sample(count_1)
df_under = pd.concat([df_values_0_under,df_values_1],axis = 0)
```

```python
X = df_under.drop(['Exited'],axis = 'columns')
y = df_under.Exited
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=.21,stratify=y)
```

```python
y_preds = ANN(X_train,y_train,X_test,y_test,'binary_crossentropy')
```

### Oversampling

```python
df_values_1_over = df_values_1.sample(count_0,replace =True)
df_over = pd.concat([df_values_0,df_values_1_over],axis = 0)
X = df_over.drop(['Exited'],axis = 'columns')
y = df_over.Exited
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0,test_size=.21,stratify=y)
preds = ANN(X_train,y_train,X_test,y_test,'binary_crossentropy')
```

### SMOTE

```python
X = df.drop(['Exited'],axis = 'columns')
y = df.Exited
from imblearn.over_sampling import SMOTE
smote =SMOTE(sampling_strategy = 'minority')
X_sm,y_sm = smote.fit_resample(X,y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_sm,y_sm,random_state=0,test_size=.21,stratify=y_sm)
ANN(X_train,y_train,X_test,y_test,'binary_crossentropy')
```
