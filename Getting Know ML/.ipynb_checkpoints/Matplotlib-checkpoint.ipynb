### Pyplot API vs OO API
* Pyplot API : quickly
* OO API : Advance

Pyplot API

```python
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),color="red",linestyle="dashed",label="Sin X");
plt.plot(x,np.cos(x),color="blue",label="cos x");
plt.xlabel("X")
plt.ylabel("F(x)")

# plt.axis([0,4,-0.7,0.7])      
#([xmin,xmax  ,  ymin,ymax])

plt.axis('equal')
plt.legend();
plt.show();
```

## Object-Oriented API 

```python
x=[1,2,3,4]
y=[11,22,33,44]

#2. Setup plot
fig,ax=plt.subplots(figsize=[5,5])
ax.plot(x,y);

ax.set(title="A simple plot",xlabel="a-axis",ylabel='y-axis');
```

### Scatter plot 

```python
#pyplot API
plt.scatter(x,np.exp(x));
```

```python
#OO API
#Prepare Data
rng=np.random.RandomState(0)

x=rng.randn(100)
y=rng.randn(100)

colors=rng.rand(100)
sizes=1000*rng.rand(100)

fig,ax=plt.subplots()
img1=ax.scatter(x,y,s=sizes,c=colors,cmap='viridis',alpha=0.3);
plt.grid(True)
fig.colorbar(img1);
```

### Bar
*Vertical 

*Horizontal

```python
#Prepare Data from a python dictionary
soft_drink_prices={'Cokes':10,
                  'Pepsi':12,
                  'Sprite':8}

fig,ax=plt.subplots()
ax.bar(soft_drink_prices.keys(),soft_drink_prices.values());
ax.set(title="Bach hoa xanh",ylabel="Prices");
```

```python
fig,ax=plt.subplots()
ax.barh(list(soft_drink_prices.keys()),list(soft_drink_prices.values())); #convert sang list 
ax.set(xlabel='prices')
```

###  Histogram

```python
# Prepare Data
np.random.seed(42)
student_height=np.random.normal(170,10,250)
```

```python
student_height
```

```python
#Pyplot API

plt.hist(student_height,bins=30);
```

###  Subplots
* Multiple plots on one figure

```python
#Option 1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5));
#plot data to each axis
ax1.plot(x,x/2)
ax1.plot(x,x*2)

ax2.scatter(np.random.random(10),np.random.random(10));

ax3.bar(soft_drink_prices.keys(),soft_drink_prices.values())

ax4.hist(student_height,bins=10)
```

```python
# Option 2
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5));
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(soft_drink_prices.keys(),soft_drink_prices.values());
ax[1,1].hist(student_height,bins=30);
```




# Project Data Visualization -  California City 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#ladtitude : vi do
#longtitude: kinh do

cities=pd.read_csv('california_cities.csv')
cities.head(20)

cities['area_total_km2'].sort_values(ascending=False)
```

```python
#Extract Latd and Longd
lat,lon=cities['latd'],cities['longd']
population,area=cities['population_total'],cities['area_total_km2']


#Plot using Pyplot API
plt.style.use('seaborn')
plt.scatter(lon,lat,c=np.log10(population),cmap='viridis',s=area,
           linewidths=0,alpha=0.6);

plt.axis('equal');
plt.title("California Cities: Population and Area Distribution")
plt.xlabel('longtitude');
plt.ylabel('latitude');

plt.colorbar(label='log$_{10}$(population)');
plt.clim(3,7);


#create a legend for cities'size
area_range=[50,100,300,500]
for area in area_range:
    plt.scatter([],[],s=area,c='k',alpha=0.4,
                label=str(area)+'km$^2$');
plt.legend(labelspacing=1,scatterpoints=1,title="Cities Area");
```
